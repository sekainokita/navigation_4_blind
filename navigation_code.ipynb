```python
from ultralytics import YOLO
import cv2
import numpy as np
import onnxruntime

# 🔹 YOLO ONNX 모델 로드
model_1 = YOLO("/content/drive/MyDrive/navigation/block_best.onnx", task="detect")  # 점자 블록 탐지 모델
model_2 = YOLO("/content/drive/MyDrive/navigation/obstacle_test.onnx", task="detect")  # 장애물 탐지 모델

# 🔹 깊이 측정 ONNX 모델 로드 (GPU 사용 설정 추가)
depth_model = onnxruntime.InferenceSession(
    "/content/drive/MyDrive/navigation/midas_small.onnx",
    providers=["CUDAExecutionProvider", "CPUExecutionProvider"]
)

print("[INFO] ONNX Runtime Providers:", depth_model.get_providers())


def preprocess_depth_image(image, size=(128, 128)):
    img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) / 255.0
    img = cv2.resize(img, size)
    img = img.transpose(2, 0, 1).astype(np.float32)
    img = np.expand_dims(img, axis=0)
    return img

def estimate_depth(image):
    input_tensor = preprocess_depth_image(image)
    input_name = depth_model.get_inputs()[0].name
    depth_map = depth_model.run(None, {input_name: input_tensor})[0]
    depth_map = np.squeeze(depth_map)
    depth_map = cv2.resize(depth_map, (image.shape[1], image.shape[0]))
    return depth_map

# 🔹 동영상 파일 열기
video_path = "/content/drive/MyDrive/navigation/KakaoTalk_20250320_144019708.mp4"  # 사용할 동영상 경로
cap = cv2.VideoCapture(video_path)

# 🔹 비디오 저장 설정 (출력 프레임 수동 설정: 30fps), 원본 프레임 절반 크기
frame_width = int(cap.get(3) // 2)
frame_height = int(cap.get(4) // 2)
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(
    "/content/drive/MyDrive/navigation/output_with_alerts.mp4",
    fourcc,
    30.0,
    (frame_width, frame_height)
)

# 🔹 출력 창 크기 설정 (비율 유지)
display_scale = 0.5  # 50% 크기로 축소

frame_total = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
frame_index = 0

while frame_index < frame_total:
    cap.set(cv2.CAP_PROP_POS_FRAMES, frame_index)
    ret, frame = cap.read()
    if not ret:
        break

    print(f"[INFO] Processing frame {frame_index} / {frame_total}")
    frame_index += 2  # 🔹 두 프레임씩 건너뛰어 절반만 처리

    results_1 = list(model_1.predict(frame, imgsz=640, conf=0.5, stream=True))
    results_2 = list(model_2.predict(frame, imgsz=640, conf=0.5, stream=True))

    # 🔹 깊이맵 추정 (화면에는 표시하지 않음)
    depth_map = estimate_depth(frame)

    # 🔹 깊이 컬러맵 생성 (주석처리: 나중에 시각화하고 싶을 때 사용)
    # depth_colormap = cv2.applyColorMap(cv2.convertScaleAbs(depth_map, alpha=255.0 / depth_map.max()), cv2.COLORMAP_INFERNO)

    for r in results_1 + results_2:
        for box in r.boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            conf = box.conf[0]
            cls_id = int(box.cls[0])
            class_name = r.names[cls_id] if cls_id < len(r.names) else "Unknown"

            color = (0, 255, 0) if r in results_1 else (0, 0, 255)
            label_text = f"{class_name} {conf:.2f}"

            # 🔹 깊이 기반 경고 판단 (계속 활성화)
            if r in results_2:
                cx = int((x1 + x2) / 2)
                cy = int((y1 + y2) / 2)
                depth_value = depth_map[cy, cx]
                # print(f"[DEBUG] {class_name} depth value at ({cx},{cy}): {depth_value:.3f}")
                if depth_value > 400:
                    warning_text = "Danger"
                    color = (0, 0, 255)
                elif depth_value < 100:
                    warning_text = "Warning"
                    color = (0, 165, 255)
                else:
                    warning_text = "Safe"
                    color = (0, 255, 0)
                label_text += f" | {warning_text}"

            cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)
            cv2.putText(frame, label_text, (x1, y1 - 10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.6, color, 2)

    # 🔹 프레임 크기를 절반으로 줄여 저장
    resized_frame = cv2.resize(frame, (frame_width, frame_height))
    out.write(resized_frame)

    # 🔹 비율 유지하면서 화면 축소해서 출력 (주석처리: 나중에 실시간으로 보고 싶을 때 해제)
    # resized = cv2.resize(frame, (0, 0), fx=display_scale, fy=display_scale)
    # cv2.imshow("YOLO Object Detection (Video)", resized)

    # if cv2.waitKey(1) & 0xFF == 27:
    #     break

cap.release()
out.release()

# 🔹 Colab 등 GUI 없는 환경에서는 destroyAllWindows() 제거 (에러 방지)
# cv2.destroyAllWindows()
```

```python
from fastapi import FastAPI, Request
from fastapi.responses import HTMLResponse
from fastapi.staticfiles import StaticFiles
import cv2, numpy as np, base64
from io import BytesIO
import onnxruntime
from ultralytics import YOLO

app = FastAPI()
app.mount("/static", StaticFiles(directory="static"), name="static")

# 모델 로딩
model_1 = YOLO(block_model_path, task="detect")
model_2 = YOLO(obstacle_model_path, task="detect")
depth_model = onnxruntime.InferenceSession(
    depth_model_path,
    providers=["CUDAExecutionProvider", "CPUExecutionProvider"]
)

model_1_names = ['Go_Forward', 'Stop']
model_2_names = ['barricade', 'bench', 'bicycle', 'bollard', 'bus', 'car', 'carrier', 'chair', 'dog', 'fire_hydrant', 'kiosk', 'motorcycle', 'movable_signage', 'parking_meter', 'person', 'pole', 'potted_plant', 'scooter', 'stop', 'stroller', 'table', 'traffic_light', 'traffic_sign', 'tree_trunk', 'truck', 'wheelchair']

def preprocess_depth_image(image, size=(128, 128)):
    img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) / 255.0
    img = cv2.resize(img, size)
    img = img.transpose(2, 0, 1).astype(np.float32)
    return np.expand_dims(img, axis=0)

def estimate_depth(image):
    input_tensor = preprocess_depth_image(image)
    input_name = depth_model.get_inputs()[0].name
    depth_map = depth_model.run(None, {input_name: input_tensor})[0]
    return cv2.resize(np.squeeze(depth_map), (image.shape[1], image.shape[0]))

def process_frame(frame: np.ndarray) -> np.ndarray:
    result_1 = model_1.predict(frame, imgsz=640, conf=0.5)[0]
    result_2 = model_2.predict(frame, imgsz=640, conf=0.5)[0]
    depth_map = estimate_depth(frame)
    depth_max = depth_map.max() if depth_map.max() > 0 else 1.0

    for r, names, base_color in [(result_1, model_1_names, (0,255,0)), (result_2, model_2_names, (0,0,255))]:
        for box in r.boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            conf = float(box.conf[0])
            cls_id = int(box.cls[0])
            class_name = names[cls_id] if cls_id < len(names) else "Unknown"
            label = f"{class_name} {conf:.2f}"
            color = base_color

            if r == result_2:
                cx, cy = (x1 + x2)//2, (y1 + y2)//2
                depth_val = depth_map[cy, cx]
                if depth_val > 400:
                    label += " | Danger"
                    color = (0, 0, 255)
                elif depth_val < 100:
                    label += " | Warning"
                    color = (0, 165, 255)
                else:
                    label += " | Safe"

            cv2.rectangle(frame, (x1,y1), (x2,y2), color, 2)
            cv2.putText(frame, label, (x1, y1-10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

    return frame

@app.get("/", response_class=HTMLResponse)
def index():
    return open("static/index.html", "r", encoding="utf-8").read()

@app.post("/process_frame")
async def process_frame_api(request: Request):
    data = await request.json()
    img_data = data["image"].split(",")[1]
    img_bytes = base64.b64decode(img_data)
    frame = cv2.imdecode(np.frombuffer(img_bytes, np.uint8), cv2.IMREAD_COLOR)
    result_img = process_frame(frame)
    _, buffer = cv2.imencode(".jpg", result_img)
    return {"image": base64.b64encode(buffer).decode("utf-8")}
```

```python
import cv2
import onnxruntime
import numpy as np
import os
import matplotlib.cm as cm

# 🔧 설정
onnx_path = "/content/drive/MyDrive/navigation/depth_anything_v2_vits.onnx"  # 모델 경로
video_path = "/content/drive/MyDrive/navigation/short_video.mp4"             # 입력 비디오
output_path = "/content/drive/MyDrive/navigation/depth_output.mp4"           # 출력 비디오
input_size = 518
grayscale = False

# 🔁 ONNX 실행 세션
session = onnxruntime.InferenceSession(onnx_path, providers=["CUDAExecutionProvider", "CPUExecutionProvider"])
input_name = session.get_inputs()[0].name
output_name = session.get_outputs()[0].name

# 🎞 비디오 로딩
cap = cv2.VideoCapture(video_path)
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)
frame_total = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

# 🔽 출력 설정
fourcc = cv2.VideoWriter_fourcc(*"mp4v")
out = cv2.VideoWriter(output_path, fourcc, fps, (frame_width, frame_height))

cmap = cm.get_cmap("Spectral_r")

print("🚀 추론 시작...")

for frame_idx in range(frame_total):
    ret, frame = cap.read()
    if not ret:
        break

    # 전처리
    img = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    img = cv2.resize(img, (input_size, input_size)) / 255.0
    input_tensor = img.transpose(2, 0, 1).astype(np.float32)[np.newaxis, ...]

    # 추론
    depth = session.run([output_name], {input_name: input_tensor})[0]
    depth = np.squeeze(depth)
    depth = cv2.resize(depth, (frame.shape[1], frame.shape[0]))
    depth_norm = ((depth - depth.min()) / (depth.max() - depth.min()) * 255.0).astype(np.uint8)

    # 시각화
    if grayscale:
        depth_vis = np.repeat(depth_norm[..., np.newaxis], 3, axis=-1)
    else:
        depth_vis = (cmap(depth_norm)[:, :, :3] * 255).astype(np.uint8)[..., ::-1]

    # 결합 및 저장
    # combined = cv2.hconcat([frame, depth_vis])
    out.write(depth_vis)

    print(f"\rFrame {frame_idx + 1}/{frame_total}", end="")

cap.release()
out.release()

print("\n✅ 완료! 출력 저장:", output_path)
```

# midas 하이브리드로 동영상 테스트

```python
input_video_path = "/content/drive/MyDrive/navigation/short_video.mp4"
output_video_path = "/content/drive/MyDrive/navigation/midas_hybrid_output.mp4"
```

```python
import cv2
import numpy as np
import os
import torch

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("🔥 Using device:", device)

# ✅ 4. MiDaS Hybrid 모델 로드
model_type = "DPT_Hybrid"
midas = torch.hub.load("intel-isl/MiDaS", model_type)
midas.to(device)
midas.eval()

# ✅ 5. 공식 전처리 로딩
transform = torch.hub.load("intel-isl/MiDaS", "transforms").dpt_transform

# ✅ 6. 비디오 로드
cap = cv2.VideoCapture(input_video_path)
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)
frame_total = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

# ✅ 7. 출력 설정
fourcc = cv2.VideoWriter_fourcc(*"mp4v")
out = cv2.VideoWriter(output_video_path, fourcc, fps, (frame_width, frame_height))

print("🚀 MiDaS Hybrid로 영상 처리 시작...")

for frame_idx in range(frame_total):
    ret, frame = cap.read()
    if not ret:
        break

    print(f"\rFrame {frame_idx + 1}/{frame_total}", end="")

    # 전처리
    img = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    input_tensor = transform(img).to(device)

    # 추론
    with torch.no_grad():
        depth = midas(input_tensor)

    depth_map = depth.squeeze().cpu().numpy()
    depth_map = cv2.resize(depth_map, (frame.shape[1], frame.shape[0]))

    # 정규화 및 컬러맵
    depth_norm = ((depth_map - depth_map.min()) / (depth_map.max() - depth_map.min()) * 255.0).astype(np.uint8)
    depth_colormap = cv2.applyColorMap(depth_norm, cv2.COLORMAP_INFERNO)

    # 저장
    out.write(depth_colormap)

cap.release()
out.release()
print("\n✅ 완료! 저장된 파일:", output_video_path)
```

## 🧭 시각장애인 보행 보조 시스템 - 기능 정리

YOLO 객체 탐지 + Depth Anything V2 거리 추정을 기반으로 하는  
**실시간 안내 시스템**의 주요 기능 및 우선순위 정리입니다.

---

### ⚙️ 전체 구조 요약

- **YOLO v8 ONNX 모델 2개 사용**
  - `block_best.onnx`: 점자블록 탐지 (`Go_Forward`, `Stop`)
  - `obstacle_test.onnx`: 장애물 탐지
- **Depth Anything V2 Small ONNX 사용**
  - 거리 추정용 (`depth_anything_v2_vits.onnx`)
- **출력**
  - 영상 프레임 위에 안내 메시지 실시간 출력
  - 우선순위에 따라 안내가 충돌되지 않도록 제어

---

### 📋 기능 목록 및 우선순위

| 순위 | 기능 이름 | 조건 | 동작 | 출력 메시지 |
|------|-----------|--------|--------|---------------|
| 🥇 1 | **정지 안내 (STOP)** | 클래스가 `"stop"`이고, `y2 > frame_height * 0.95` | 3초간 다른 안내 차단 | `Stop sign detected! Please stop.` |
| 🥈 2 | **장애물 회피 안내** | 깊이 추정값 기준<br>`depth > 0.4`, `cx` 좌우 위치 | 장애물 회피 유도 | `Obstacle close! Move left.` / `Move right.` |
| 🥉 3 | **회전 안내** | 점자블록이 **가로형** (`w/h > 1.5`)이고, `cy > frame_height * 0.6` | 꺾인 점자블록 따라가도록 안내 | `Turn Left` / `Turn Right` |
| 4 | **점자블록 이탈 안내** | 점자블록 중심이 좌/우 30% 밖에 있음 | 복귀 유도 | `Off the guide block. Move left/right.` |
| 5 | **점자블록 미감지 안내** | 점자블록이 아예 없음 | 사용자에게 경고 | `No guide block detected` |

---

### 정지 안내 (stop 점자블록)

- `stop_detected_time` 변수로 **최초 Stop 감지 시간** 저장
- 이후 **3초간** 다른 안내 (`Turn`, `Obstacle`, `Guide block`) 출력 **차단**
- 3초가 지나면 다른 안내 재개
- 정지 바운딩 박스가 화면 하단 (전체 화면의 5%)부분에 닿으면 멈추라고 출력

### 장애물 회피 안내

- 장애물이 danger 로 탐지되면 그 탐지된 장애물의 바운딩 박스의 중심값을 보고 그 중심점이 화면의 좌측에 위치하면 오른쪽으로 피하세요 출력, 우측에 위치하면 왼쪽으로 피하세요 출력

### 회전 안내
- 일반적으로 직진할때는 세로로 긴 점자블록이 탐지되지만 점자 블록이 꺾였을 경우. 즉, 좌회전이나 우회전을 회야되는 경우는 가로로 긴 점자 블록이 탐지될 것이다.
-가로가 세로보다 긴 Go_Foward 점자 블록이 탐지되면 그 점자블록의 중앙값을 보고 중심점이 화면의 좌측에 위치하면 왼쪽으로 회전하세요 출력, 우측에 위치하면 오른쪽으로 회전하세요 출력

### 점자블록 이탈 안내
- 직진을 하다가 점자블록을 차츰 벗어나는 경우. 탐지되는 점자블록의 중앙값이 화면의 우측 30% 혹은 좌측 30%에 들어갈 경우 안내문 출력

### 점자블록 미감지 안내
- 점자블록이 미감지 될 경우 출력. 보통 점자블록의 상태가 안좋거나 회전하는 경우 발생

```python
import cv2
import numpy as np
import onnxruntime
import time
from ultralytics import YOLO

# 항상 메시지 출력 시간을 위해 시간 보관 모델
stop_detected_time = None
stop_hold_seconds = 3

# ▪️ 모델 로드
model_1 = YOLO("/content/drive/MyDrive/navigation/block_best.onnx", task="detect")
model_2 = YOLO("/content/drive/MyDrive/navigation/obstacle_test.onnx", task="detect")

# ▪️ Depth Anything V2 Small ONNX
depth_model = onnxruntime.InferenceSession(
    "/content/drive/MyDrive/navigation/depth_anything_v2_vits.onnx",
    providers=["CUDAExecutionProvider", "CPUExecutionProvider"]
)

def preprocess_depth_image(image, size=(518, 518)):
    img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) / 255.0
    img = cv2.resize(img, size)
    img = img.transpose(2, 0, 1).astype(np.float32)
    return np.expand_dims(img, axis=0)

def estimate_depth(image):
    input_tensor = preprocess_depth_image(image)
    input_name = depth_model.get_inputs()[0].name
    depth_map = depth_model.run(None, {input_name: input_tensor})[0]
    depth_map = np.squeeze(depth_map)
    depth_map = cv2.resize(depth_map, (image.shape[1], image.shape[0]))
    depth_map = (depth_map - depth_map.min()) / (depth_map.max() - depth_map.min())
    return depth_map

video_path = "/content/drive/MyDrive/navigation/long_video.mp4"
cap = cv2.VideoCapture(video_path)

frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)
max_frames = int(fps * 60)

fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter("/content/drive/MyDrive/navigation/output_depthanything_vits.mp4", fourcc, fps, (frame_width, frame_height))

print("\ud83d\ude80 YOLO + Depth Anything v2")

frame_index = 0
while frame_index < max_frames:
    ret, frame = cap.read()
    if not ret:
        break

    print(f"[INFO] Processing frame {frame_index + 1} / {max_frames}")
    frame_index += 1

    results_1 = list(model_1.predict(frame, imgsz=640, conf=0.5, stream=True))
    results_2 = list(model_2.predict(frame, imgsz=640, conf=0.5, stream=True))
    depth_map = estimate_depth(frame)

    block_detected = False
    stop_detected = False
    skip_other_guidance = False

    for r in results_1 + results_2:
        for box in r.boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            conf = float(box.conf[0])
            cls_id = int(box.cls[0])
            class_name = r.names[cls_id] if cls_id < len(r.names) else "Unknown"

            color = (0, 255, 0) if r in results_1 else (0, 0, 255)
            label_text = f"{class_name} {conf:.2f}"

            if r in results_1:
                block_detected = True
                cx = int((x1 + x2) / 2)
                cy = int((y1 + y2) / 2)
                w = x2 - x1
                h = y2 - y1
                ratio = w / h

                # Stop block 추가
                if class_name.lower() == "stop" and y2 > frame_height * 0.95:
                    stop_detected = True
                    if stop_detected_time is None:
                        stop_detected_time = time.time()

                    cv2.putText(frame, "Stop sign detected! Please stop.",
                                (50, frame_height - 150), cv2.FONT_HERSHEY_SIMPLEX,
                                1.2, (0, 0, 255), 3, cv2.LINE_AA)
                    skip_other_guidance = True

            if stop_detected_time:
                if time.time() - stop_detected_time < stop_hold_seconds:
                    skip_other_guidance = True
                else:
                    skip_other_guidance = False
                    stop_detected_time = None

            if not skip_other_guidance:
                if r in results_1:
                    left_threshold = int(frame_width * 0.3)
                    right_threshold = int(frame_width * 0.7)

                    if cx < left_threshold:
                        cv2.putText(frame, "Off the guide block. Move right.",
                                    (50, frame_height - 60), cv2.FONT_HERSHEY_SIMPLEX,
                                    1.0, (255, 255, 0), 3, cv2.LINE_AA)
                    elif cx > right_threshold:
                        cv2.putText(frame, "Off the guide block. Move left.",
                                    (50, frame_height - 60), cv2.FONT_HERSHEY_SIMPLEX,
                                    1.0, (255, 255, 0), 3, cv2.LINE_AA)

                    if ratio > 1.5 and cy > frame_height * 0.6:
                        if cx < frame_width * 0.5:
                            cv2.putText(frame, "Turn Left",
                                        (50, frame_height - 100), cv2.FONT_HERSHEY_SIMPLEX,
                                        1.2, (0, 0, 255), 3, cv2.LINE_AA)
                        else:
                            cv2.putText(frame, "Turn Right",
                                        (50, frame_height - 100), cv2.FONT_HERSHEY_SIMPLEX,
                                        1.2, (0, 0, 255), 3, cv2.LINE_AA)

                if r in results_2:
                    cx = int((x1 + x2) / 2)
                    cy = int((y1 + y2) / 2)
                    depth_value = depth_map[cy, cx]

                    if depth_value > 0.4:
                        warning_text = "Danger"
                        color = (0, 0, 255)
                        if cx < frame_width * 0.5:
                            cv2.putText(frame, "Obstacle close! Move right.",
                                        (50, frame_height - 120), cv2.FONT_HERSHEY_SIMPLEX,
                                        1.0, (0, 0, 255), 3, cv2.LINE_AA)
                        else:
                            cv2.putText(frame, "Obstacle close! Move left.",
                                        (50, frame_height - 120), cv2.FONT_HERSHEY_SIMPLEX,
                                        1.0, (0, 0, 255), 3, cv2.LINE_AA)
                    elif depth_value > 0.2:
                        warning_text = "Warning"
                        color = (0, 165, 255)
                    else:
                        warning_text = "Safe"
                        color = (0, 255, 0)

                    label_text += f" | {warning_text}"

            cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)
            cv2.putText(frame, label_text, (x1, y1 - 10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.6, color, 2)

    if not block_detected:
        cv2.putText(frame, "No guide block detected",
                    (50, frame_height - 60), cv2.FONT_HERSHEY_SIMPLEX,
                    1.0, (0, 0, 255), 3, cv2.LINE_AA)

    out.write(frame)

cap.release()
out.release()
print("\u2705 Done: output saved")
```
